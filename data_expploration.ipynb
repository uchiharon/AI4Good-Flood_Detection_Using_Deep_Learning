```python
from ipyfilechooser import FileChooser
from IPython.display import Image, display
import ipywidgets as widgets

# Create and display a FileChooser widget
fc = FileChooser('./selected_data/train/images/')
fc.filter_pattern = '*pre*'
display(fc)

# Function to display the selected image
def display_selected_image(chooser):
    selected_file = chooser.selected
    if selected_file:
        display(Image(filename=selected_file))

# Button to confirm the file selection and display the image
confirm_button = widgets.Button(description="Display Image")
display(confirm_button)

# Event handler for the button
def on_button_clicked(b):
    display_selected_image(fc)

confirm_button.on_click(on_button_clicked)
```
